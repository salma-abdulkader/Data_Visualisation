# **Plot Fig**

# **Scatter Plot**

```python
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [5, 2, 4, 2, 1, 4, 5, 2]
plt.scatter(x, y, label = "skitscat", color = "blue", marker = "*", s = 200)  # s -> marker_size
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('x-axis vs y-axis')
plt.legend()
plt.show()
```

```python
plt.scatter(data['EstimatedSalary'], data['Age'], label = "new", color = "green", marker = "*", s = 20, alpha = 0.5)
plt.xlabel('EstimatedSalary')
plt.ylabel('Age')
plt.title('EstimatedSalary vs Age')
plt.legend()
plt.show()

# Controls transparency of the plotted objects. It takes a value between 0 and 1, where:
# 0: Completely transparent (invisible)
# 1: Completely opaque (solid)
```

```python
plt.scatter(data['Age'], data['EstimatedSalary'], label = "old", color = "green", marker = "*", s = 20)
plt.scatter(data['Age'], data['User ID'], label = "new", color = "red", marker = ">", s = 20)
plt.xlabel('age')
plt.ylabel('thalachh')
plt.title('age vs thalachh')
plt.legend()
plt.show()
```

# **Stack Plot**

```python
days = [1, 2, 3, 4, 5]
sleeping = [7, 8, 6, 11, 7]
eating = [2, 3, 4, 3, 2]
working = [7, 8, 7, 2, 2]
playing = [8, 5, 7, 8, 13]
plt.stackplot(days, sleeping, eating, working, playing, colors = ['purple', 'blue', 'pink', 'yellow'])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('x-axis vs y-axis')
plt.legend()
plt.show()
```

```python
days = [1, 2, 3, 4, 5]
sleeping = [7, 8, 6, 11, 7]
eating = [2, 3, 4, 3, 2]
working = [7, 8, 7, 2, 2]
playing = [8, 5, 7, 8, 13]

plt.plot([], [], color = 'purple', label = 'Eating', linewidth = 5)
plt.plot([], [], color = 'blue', label = 'Sleeping', linewidth = 5)
plt.plot([], [], color = 'pink', label = 'Eating', linewidth = 5)
plt.plot([], [], color = 'yellow', label = 'Sleeping', linewidth = 5)

plt.stackplot(days, sleeping, eating, working, playing, colors = ['purple', 'blue', 'pink', 'yellow'])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('x-axis vs y-axis')
plt.legend()
plt.show()
```

# **SubPlots**

```python
x = np.arange(100)
fig, axs = plt.subplots(2, 2)

axs[0,0].plot(x, np.sin(x), color = "purple")
axs[0,0].set_title('Sine')

axs[0,1].plot(x, np.cos(x), color = "pink")
axs[0,1].set_title('Cosine')

axs[1,0].plot(x, np.tan(x), color = "yellow")
axs[1,0].set_title('Tangent')

axs[1,1].plot(x, np.exp(x), color = "cyan")
axs[1,1].set_title('Exponential')
axs[1,1].set_xlabel("Tesst")
axs[1,1].set_ylabel("Tessssst")

fig .suptitle("Four Plots")
plt.tight_layout()
plt.show()
```
